```python
import pandas as pd

# Setup
print("Lung Cancer Overview: ")
df = pd.read_csv("lung_cancer_prediction_dataset.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# 1. Create a Boxplot of Lung Cancer Deaths Distribution
plt.figure(figsize=(8, 6))
plt.boxplot(df['Annual_Lung_Cancer_Deaths'])
plt.title('Distribution of Annual Lung Cancer Deaths Across Countries')
plt.xlabel('All Countries')
plt.ylabel('Annual Deaths per Country')
plt.show()
```

```python
# 3. Create a Density Plot of the Lung Cancer Mortality Rate.

import seaborn as sns

sns.kdeplot(df["Mortality_Rate"], bw_method = 0.5)

plt.title("Density Plot of Lung Cancer Mortality Rate")
plt.xlabel("Mortality Rate")
plt.ylabel("Density")

plt.show()
```

```python
# Setup
print("Global Air Pollution Overview: ")
df = pd.read_csv("global_air_pollution_dataset.csv")
print(df.head())
```

```python
# 2. Create a Histogram of PM2.5 AQI Values.

plt.figure(figsize=(10, 6))
plt.hist(df['PM2.5_AQI_Value'], bins=30)
plt.title('Histogram of PM2.5 AQI Values')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
# Part 4

# 1. Use the matplotlib library for this graph.

# Setup
df = pd.read_csv("global_air_pollution_dataset.csv")

pm25_data = df['PM2.5_AQI_Value']

sns.histplot(
    data=pm25_data,
    stat='density',
    color='purple',
    bins=30
)

sns.kdeplot(
    data=pm25_data,
    color='red',
    linewidth=2,
    fill = True,
    alpha = 0.4
)

plt.title('PM2.5 AQI Distribution with Density Overlay')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Density')

plt.show()
```
